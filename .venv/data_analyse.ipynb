```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
run1 = pd.read_csv("data/run1/fits.csv")
run2 = pd.read_csv("data/run2/fits.csv")
run3 = pd.read_csv("data/run3/fits.csv")
run4 = pd.read_csv("data/run4/fits.csv")
run5 = pd.read_csv("data/run5/fits.csv")

# Create a single line chart comparing the fitness values from each run
plt.plot(run1['fitness'], label='Run 1')
plt.plot(run2['fitness'], label='Run 2')
plt.plot(run3['fitness'], label='Run 3')
plt.plot(run4['fitness'], label='Run 4')
plt.plot(run5['fitness'], label='Run 5')
plt.xlabel('Iteration')
plt.ylabel('Fitness')
plt.yticks(np.arange(30, 70, 10))
plt.title('Fitness Comparison Across Runs')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
run1 = pd.read_csv("advanced_cw_data/run1/fits.csv")
run2 = pd.read_csv("advanced_cw_data/run2/fits.csv")
run3 = pd.read_csv("advanced_cw_data/run3/fits.csv")
run4 = pd.read_csv("advanced_cw_data/run4/fits.csv")

# Create a line chart comparing the fitness values from each run 
plt.plot(run1['mean fitness'], label='Run 1')
plt.plot(run2['mean fitness'], label='Run 2')
plt.plot(run3['mean fitness'], label='Run 3')
plt.plot(run4['mean fitness'], label='Run 4')
plt.xlabel('Iteration')
plt.ylabel('Mean Fitness')
plt.yticks(np.arange(0, 30, 10))
plt.title('Mean Fitness Comparison Across Runs')
plt.legend()
plt.show()
```
